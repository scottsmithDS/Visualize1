```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

CityData = pd.read_csv("../cities.csv")
```

```python
CityData = pd.DataFrame(CityData)
```

```python
lat = CityData['Lat']
MaxTemp = CityData['Max Temp']
WindSpeed = CityData['Wind Speed']
Cloudiness = CityData['Cloudiness']
Humidity = CityData['Humidity']
```

```python
plt.scatter(lat, Humidity, marker="o", facecolors="red", edgecolors="black")
plt.title("Latitude & Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.savefig("Lat&Humidity.png")
```
